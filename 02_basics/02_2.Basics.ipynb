```python
range(5)
# it is giving us a range of numbers between 1 and 5
# it is very important to know that the last number of the range will not get printed it is excluded
```

```python
# for loop
for i in range(5):
    print(i) # we are printing the numbers from 0 to 4, the last number is not going to get printed
    # but the total numbers of elements are five
```

```python
for i in range(1,6):  #  the starting point is 1 and the ending point is 5 because the range function excludes the last number
    print(i)
```

```python

for i in range(1,10,1): # syntax is start, end-1 and steps (three parameters)
    print(i)
    # we are printing the numbers from 1 to 9 with steps of 1
```

```python
for i in range(10,1,-1):
    print(i) # we are printing the numbers in reverse order from 10 to 2 with a step size or decrement of -1
```

```python
for i in range(10,2,-2): # we are printing the numbers in reverse order from 10 to 2 with a step size or decrement of -2
    print(i)
```

```python
# strings

name = "Shravan"

for i in name:
    print(i)
```

```python
## while loop: executes as long as the condition is true
# until the condition is false the loop keeps on executing
count  = 0
while count < 5:
    print(count)
    count = count+1
```

```python
count  = 0

while count%2 == 0: # 0 mod 2 is 0, and then 1 mod 2 is 1 which is false, so it comes out of the loop
    print(count)
    count = count+1
```

```python
# Print the sum of first N natural numbers
num = int(input("Enter a number:"))
total = 0
for i in range(1,num+1):
    total = total + i
print("sum = " , total)
```

```python
# print even numbers between 1 and 50
for i in range(2 , 51 , 2):
    print(i)
```

```python
# Print factorial of a number

num = int(input("Enter a number:"))
fact = 1

for i in range(1,num+1):
    fact = fact*i
print(fact)
```

```python
## Loop control statements

## break
## break statement exits the loop prematurely
## break statement

for i in range(10): # we are starting from the 0 to 9
    if i == 5: # if i == 5 then we will come out of the loop
        break
    print(i)

    ## only when i == 5 do the break and come outside the for loop
```

```python
## continue: skips the current iteration and continues the next

## simple understanding of the below example
""" we are iterating over 0 to 9 and if the i%2 == 0 which means even then continue
    and if it is not then print  i which will be the odd numbers so we get odd numbers 1,3,5,7,9
"""

for i in range(10):
    if i%2 == 0:
        continue
    print(i)
```

```python
## Pass: is a null operation it does nothing

for i in range(5):
    if i == 3:
        pass # it simply skips the if block
    print(i)
```

```python
## Nested loops: Loop inside loop is nested loop

for i in range(3): # runs from 0 to 2
    for j in range(2): # runs from 0 to 1
        print(f"i:{i} and j: {j}")

# simple explanation for every outer loop , the inner loop executes completely
"""
outer loop | inner loop
     0     |  (0,1)executes
     1     |  (0,1)executes
     2     |  (0,1)executes

and so is the output
"""
```

```python
## examples - calculate the sum of first N natural numbers using a while and for loop both

#1 : while loop
num = int(input("Enter a number:")) # user inputs: 5
result = 0
count = 1
while count <=num :
    result = result + count
    count = count + 1
print(f"The sum of first {num} is:",result)

# we are taking two variables which are result and count and setting it to 0 and 1 respectively
# we will start the count from one because all natural numbers start from 1
# using while loop we will traverse from 1 to the number which is entered by the user,
# and we are also making sure that the result gets added with the incrementing count value
```

```python
# 2. for loop

num = int(input("Enter a number:"))
result = 0
for i in range(1,num+1):
    result = result + i
print(f"The sum of first {num} is:",result)
```

```python
## examples - prime numbers between 1 and 100

for num in range(1,101):
    if num>1:
        for i in range(2,num):
            if num % i == 0:
                break
        else:
            print(num)


```
