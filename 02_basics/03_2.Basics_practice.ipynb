```python
# Problem 1: Print a rectangle of * of size 4 rows and 6 columns.
"""
******
******
******
******

"""
for i in range(1, 5):
    for j in range(1, 7):
        print("*", end="")
    print("")

"""
logic : there are 4 rows which are marked by the i variable and 6 columns which are marked by the j variable
        for every iteration of the outer loop or the row , the inner loop or the column will execute 6 times
        so the outer loop will execute 4 times and for each time the inner loop will execute 6 times
"""
```

```python
# Problem 2: Print a right-angle triangle

"""
*
**
***
****
*****

"""

for i in range(1, 6):
    for j in range(i):
        print("*", end="")
    print()

    """

    Logic: i points the rows and j points the columns, the outer loop executes i times and based on the value of i the inner loop executes j times eg if the outer loop is 1 then inner loop will execute 1 time , if it is 2 then inner loop will execute 2 times if it is 3 then inner loop will execute 3 times and if it is 4 then inner loop will execute 4 times ans so on

    """
```

```python
# Problem 3: Number pattern (increasing numbers)

"""
1
12
123
1234
12345

"""

for i in range(1, 6):
    for j in range(1,i+1):
        print(j,end="")
    print()


"""
Logic:

when i = 1 , j = iterates from (1,1+1) = 1
when i = 2 , j = iterates from (1,2+1) = 1 2
when i = 3 , j = iterates from (1,3+1) = 1 2 3
when i = 4 , j = iterates from (1,4+1) = 1 2 3 4
when i = 5 , j = iterates from (1,5+1) = 1 2 3 4 5

"""

```

```python
# Problem 4: Number pattern (row repeated)

"""
1111
2222
3333
4444
"""
for i in range(1,5):
    for j in range(1,5):
        print(i,end="")
    print()


"""
Logic:
when i = 1 , j = iterates 4 times, and we are printing i which is 1 , which will be printed 4 times
when i = 2 , j = iterates 4 times, and we are printing i which is 2 , which will be printed 4 times
and so on
"""

```

```python
# problem 5: Multiplication Table (nested loops)
# Print multiplication table 1 to 5.

for i in range(1,6):
    for j in range(1,6):
        print(i*j,end=" ")
    print()

"""
Logic:
when
i=1 , j (1,2,3,4,5) , and logic = (i*j) = (1*1,1*2,1*3,1*4,1*5)
i=2 , j (1,2,3,4,5) , and logic = (i*j) = (2*1,2*2,2*3,2*4,2*5)
and so on
"""
```

```python
# problem 6:Hollow square pattern
"""
*****
*   *
*   *
*   *
*****

"""

for i in range(1,6):
    for j in range(1,6):
        if i==1 or i==5 or j==1 or j==5:
            print("*",end="")
        else:
            print(end=" ")
    print()


"""
Logic: there are five rows and five columns and the first and last row are filled and first and last column are filled
       and rest all the rows and columns are empty , so we are iterating over 5 rows and 5 columns , and we are keeping a
       condition that if the first row or the fourth row or the first column or the last column then we are printing * else we are printing space
"""
```
