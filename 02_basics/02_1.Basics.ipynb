```python
# What is if statement: evaluates the condition and executes the block of code within it is condition is true otherwise it
# executes the else block

age = 20
if age>=18:
    print("Eligible to vote")
else:
    print("Not eligible to vote")
```

```python
## elif statement: also the else-if statement
# allows us to check multiple conditions

marks = float(input("Enter your marks: "))

if 90 <= marks <= 100:
    print("A Grade")
elif 75<=marks<=89:
    print("B Grade")
elif 60<=marks<=74:
    print("C Grade")
elif 40<=marks<=59:
    print("D Grade")
else:
    print("Fail")

```

```python
num = int(input("Enter a number: "))
if num>0:
    print("Positive number")
    if num%2==0:
        print("Even number")
    else:
        print("Odd number")
elif num<0:
    print("Negative number")
else:
    print("The number is zero")
```

```python
# To check if the year is leap year or not
year = int(input("Enter a year: "))
if year%4==0:
    if year%100==0:
        if year%400==0:
            print(year,"Leap year")
        else:
            print(year,"Not a leap year")
    else:
        print(year,"Is a leap year")
else:
    print(year,"is not a leap year")




```

```python
# simple calculator

num1 = float(input("Enter first number"))
num2 = float(input("Enter second number"))
operation = input("Enter operation + , - , * , /")
if operation == "+":
    print(num1+num2)
elif operation == "-":
    print(num1-num2)
elif operation == "*":
    print(num1*num2)
else:
    if num2 == 0:
        print("invalid operation division by zero")
    else:
        print(num1/num2)
```

```python
# Determine ticket price based on age
age = int(input("Enter your age: "))
is_student = input("Are you a student (yes/no)").lower()

if age < 5:
    price = "free"
elif age<=12:
    price = "30 rupees"
elif age<=17:
    if is_student == "yes":
        price = "40 rupees"
    else:
        price = "50 rupees"
elif age <=64:
    if is_student == "yes":
        price = "60 rupees"
    else:
        price = "70 rupees"
else:
    price = "80 rupees"

print("Ticket price :",price)
```
