#### Magic Methods
Magic methods in python, also known as dunder methods(double underscore methods) are special methods
that start and end with double underscores. These methods enable you to define the behavior of objects
for built-in operations, such as arithmatic operations, comparison and more

Magic methods are predefined methods in python that we can use to override to change the behaviour of
your objects. Some common methods include
1. __init__: Initializes a new instance of a class
2. __str__: Returns a string representation of object
3. __repr__: returns an official string representation of an object
4. __len__: returns the length of an object
5. __getitem__: gets an item from a container
6. __setitem__: sets an item in a container

```python
class Person:
    pass
person = Person()
dir(person) ## we can see all the methods available
```

```python
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age

person = Person('Shravan' , 20)
print(person) # we get the memory location
```

```python
## using the __str__
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age

    def __str__(self):
        return f'{self.name} is {self.age} years old'


    def __repr__(self):
        return f'Person {self.name} is {self.age} years old'

person = Person('Shravan' , 20)
print(person) # we get the message in a proper format as per the overridden method
print(person.__repr__()) ## we can write it as
print(repr(person))
```
