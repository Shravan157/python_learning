#### Simple Demonstration of Library System using classes and objects


```python
class Book:
    def __init__(self, title, author, price, isbn, is_available=True):
        self.title = title
        self.author = author
        self.price = price
        self.isbn = isbn
        self.is_available = is_available

    def borrow(self):
        if self.is_available:
            self.is_available = False
            print(f'{self.title} is borrowed')
        else:
            print(f'{self.title} is not available')

    def return_book(self):
        self.is_available = True
        print(f'{self.title} is returned ')

    def get_info(self):
        return (
            f'Book title: {self.title}\n'
            f'Author: {self.author}\n'
            f'ISBN: {self.isbn}\n'
            f'Available: {self.is_available}'
        )


book1 = Book("Harry Potter", "J.K. Rowling", "9780439708180", True)
book2 = Book("The Jungle Book", "Rudyard Kipling", "9780141325293", False)

print(book1.get_info())
book1.borrow()
print(book1.get_info())

book1.return_book()
print(book1.get_info())
```

```python
class Library:
    def __init__(self, name):
        self.name = name
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def display_books(self):
        print(f"\nWelcome to {self.name} Library")
        print("Books Available:\n")
        for book in self.books:
            print(book.get_info())
            print("-" * 30)

    def find_book(self, title):
        for book in self.books:
            if book.title == title:
                return book
        return None

    def borrow_book(self, title):
        book = self.find_book(title)
        if book:
            if book.is_available:
                book.borrow()
            else:
                print(f"{title} is currently not available.")
        else:
            print(f"{title} not found in the library.")
```

```python
library = Library("City Central")

book1 = Book("Harry Potter", "J.K. Rowling", "9780439708180" , True)
book2 = Book("The Jungle Book", "Rudyard Kipling", "9780141325293", False)

library.add_book(book1)
library.add_book(book2)

library.display_books()
library.borrow_book("Harry Potter")
library.borrow_book("The Jungle Book")
library.borrow_book("Unknown Book")
```

```python
class Student:
    def __init__(self,name,roll_no,marks):
        self.name = name
        self.roll_no = roll_no
        self.marks = marks

    def display_details(self):
        return f'Student name: {self.name}\nRoll.no: {self.roll_no}\nMarks: {self.marks}'

    def is_passed(self):
        if self.marks >=40:
            status = 'Passed'
        else:
            status = 'Failed'
        return f'Status: {status}'

    def add_grace_marks(self,grace):
        self.marks += grace
        return f'Marks: {self.marks}'

    def update_marks(self,new_marks):
        self.marks = new_marks
        return f'Marks : {self.marks}'

student1 = Student('Shravan' , '28' , 90)
student2 = Student('Rahul' , '50' , 35)

print(student2.display_details())
print(student2.is_passed())
print()
print('After adding grace marks:')
print(student2.add_grace_marks(10))
print(student2.is_passed())

print(student2.update_marks(55))
```
