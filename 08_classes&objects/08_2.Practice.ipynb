#### Coding assignment on inheritance in OOPs

```python
#### Coding assignment on inheritance in OOPs

# Base class
class Person:
    def __init__(self, name, age, gender, id_number):
        self.name = name
        self.age = age
        self.gender = gender
        self.id_number = id_number

    def get_details(self):
        print("Name:", self.name)
        print("Age:", self.age)
        print("Gender:", self.gender)
        print("ID:", self.id_number)

    def update_age(self, new_age):
        self.age = new_age
        print("Updated Age:", self.age)


# Derived class: Student
class Student(Person):
    def __init__(self, name, age, gender, id_number, grade):
        super().__init__(name, age, gender, id_number)
        self.grade = grade
        self.subjects = []
        self.grades_dict = {}

    def add_subject(self, subject_name):
        if subject_name not in self.subjects:
            self.subjects.append(subject_name)
            print("Subject added successfully")
        else:
            print("Subject already exists")

    def add_grade(self, subject, grade):
        self.grades_dict[subject] = grade

    def calculate_average(self):
        if len(self.grades_dict) == 0:
            return 0
        total = sum(self.grades_dict.values())
        return total / len(self.grades_dict)

    def get_student_report(self):
        print("\n--- Student Report ---")
        self.get_details()
        print("Grade:", self.grade)
        print("Subjects:", self.subjects)
        print("Grades:", self.grades_dict)
        print("Average Marks:", self.calculate_average())


# Derived class: Teacher
class Teacher(Person):
    def __init__(self, name, age, gender, id_number, subject_taught, years_of_exp, salary):
        super().__init__(name, age, gender, id_number)
        self.subject_taught = subject_taught
        self.years_of_exp = years_of_exp
        self.salary = salary

    def give_raise(self, percentage):
        self.salary += self.salary * (percentage / 100)
        print("Updated Salary:", self.salary)

    def get_teacher_info(self):
        print("\n--- Teacher Information ---")
        self.get_details()
        print("Subject Taught:", self.subject_taught)
        print("Years of Experience:", self.years_of_exp)
        print("Salary:", self.salary)


# Base class for multiple inheritance
class Employee:
    def __init__(self, employee_id, department, salary):
        self.employee_id = employee_id
        self.department = department
        self.salary = salary

    def calculate_bonus(self):
        return (10 / 100) * self.salary


# Multiple inheritance class
class TeachingAssistant(Employee, Student):
    def __init__(self, employee_id, department, salary,
                 name, age, gender, id_number, grade, hours_per_week):

        Employee.__init__(self, employee_id, department, salary)
        Student.__init__(self, name, age, gender, id_number, grade)
        self.hours_per_week = hours_per_week

    def get_complete_profile(self):
        print("\n--- Teaching Assistant Profile ---")
        self.get_details()
        print("Grade:", self.grade)
        print("Subjects:", self.subjects)
        print("Grades:", self.grades_dict)
        print("Employee ID:", self.employee_id)
        print("Department:", self.department)
        print("Salary:", self.salary)
        print("Hours per week:", self.hours_per_week)

    def calculate_total_income(self):
        return self.salary + self.calculate_bonus()


#creating objects

student = Student("Shravan", 25, "Male", "S107", "12th")
student.add_subject("Physics")
student.add_subject("Chemistry")
student.add_grade("Physics", 88)
student.add_grade("Chemistry", 92)
student.get_student_report()

teacher = Teacher("Anita", 35, "Female", "T201", "Math", 10, 50000)
teacher.get_teacher_info()
teacher.give_raise(10)

ta = TeachingAssistant(
    employee_id="TA301",
    department="Computer Science",
    salary=30000,
    name="Sonalika",
    age=22,
    gender="Female",
    id_number="S301",
    grade="Masters",
    hours_per_week=20
)

ta.add_subject("Data Structures")
ta.add_grade("Data Structures", 95)
ta.get_complete_profile()
print("Total Income:", ta.calculate_total_income())
```
