#### Inheritance in Python
Inheritance is a fundamental concept in OOP that allows a class to inherit attributes and methods
from another class. We will discuss about single inheritance and multiple inheritance

```python
## Example of Inheritance

## Parent class
class Car:
    # instance attributes
    def __init__(self, color , model, year , engine_type):
        self.color = color
        self.model = model
        self.year = year
        self.engine_type = engine_type
    # instance methods
    def drive(self):
        print(f"Person drives the {self.engine_type} car ")
```

```python
car1 = Car('red' , '1.0',2010,'petrol')
car1.drive()
```

```python
## Child class: Single Inheritance
class Tesla(Car): ## This is how we inherit
    def __init__(self,color,model,year,engine_type,is_self_drive):
        super().__init__(color,model,year,engine_type)
        self.is_self_drive = is_self_drive
    def self_drive(self):
        print(f'Does Tesla support self driving feature? : {self.is_self_drive}')
```

```python
tesla1 = Tesla('black' , '10Gen' , '2025' , 'Electric' , True)
tesla1.self_drive()
```

```python
tesla1.drive() ## accessing the method of parent class using child class object
```

```python
## Examples of Multiple Inheritance
## When a class is inherits from more than one base class or parent class

# Base class 1
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print('Subclass must implement this method')
# Base class 2
class Pet:
    def __init__(self, owner):
        self.owner = owner

# Derived class: The class which extends both the Animal and Pet class
class Dog(Animal, Pet):
    def __init__(self, name , owner):
        Animal.__init__(self, name)
        Pet.__init__(self, owner)

    def speak(self):
        return f'{self.name} says woof'

dog1 = Dog('Tommy' , 'Shravan')
print(dog1.speak())
print(f'Owner is {dog1.owner}')
```
