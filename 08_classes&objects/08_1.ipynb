```python
## How we can define all the attributes and methods in first hand while we are initializing and creating the objects
class Dog:
    ## constructor, which takes parameters as self , and attributes/properties
    ## here name and age are some kind of attributes
    def __init__(self ,name , age):
        self.name = name # self.name is the instance variable
        self.age = age
        ## we can add any number of attributes

## creating objects
dog1 = Dog('Tommy' , 7) # we need to provide values at the runtime
dog2 = Dog('DOG',10)
print(dog1.name)
print(dog1.age)
print(dog2.name)
print(dog2.age)
```

```python
## class with instance methods
class Dog:
    def __init__(self , name , age):
        self.name = name
        self.age = age

    def bark(self):
        print(f'{self.name} says woof')

dog1 = Dog('Tommy' , 7)
dog1.bark()

dog2 = Dog('Lucy' , 10)
dog2.bark()
```

```python
## Modeling a Bank account

## Define a class for Back Account

class BankAccount:
    def __init__(self , owner , balance = 0):
        self.owner = owner
        self.balance = balance

    def deposit(self , amount):
        self.balance += amount
        print(f'{amount} is deposited. New balance is {self.balance}')
    def withdraw(self , amount):
        if amount>self.balance:
            print('Insufficient funds')
        else:
            self.balance -= amount
            print(f'{amount} is withdrawn. New balance is {self.balance}')
    def get_balance(self):
        return self.balance

## Creating objects
account = BankAccount('Shravan' , 10000)
print(account.get_balance())
```

```python
account.deposit(10000)
```

```python
account.withdraw(2000)
```
