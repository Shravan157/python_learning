```python
## Integers examples
age = 20
type(age)


## Floating point examples
temperature = 25.1
type(temperature)

## String
name = "Shravan Parthe"
type(name)

## Boolean
is_true = True
type(is_true)

print(bool()) ## by default it is false
```

```python
a = 10
b = 10
print(type(a==b))
bool(a==b)
```

```python
## common mistakes
result = "Hello" + 5  # mistake is that the Hellow is a string and 5 is an integer so we simply cannot do a concatenation
print(result)
```

```python
# Power of type conversion
result = "Hello" + str(5) ## Hence we need to convert the integer to string to perform concatenation
print(result)
```
