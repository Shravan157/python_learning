```python
"""
Introduction to variables
Declaring and Assigning variables
Naming conventions
Understanding the variable type
type checking and conversion
dynamic typing

"""

"""
Variable are fundamental elements in programming used to store the data than can be referenced and manipulated in a program
In Python variables are created when you assign a value to them and they do not need explicit declaration to reserve memory
space the declaration happens automatically when you assign a value to a variable
"""
```

```python
age = 20
height = 5.11
name = "shravan"
is_student = True

print("Age :" , age)
print("Height :" , height)
print("Name :" , name)
print("Is Student :" , is_student)
```

```python
## Naming conventions
## variable name should be meaningful
## they must always start with a letter or underscore, and can contain letter number and underscores
## variable names are case-sensitive

## valid variable name
first_name = "shravan"
last_name = "parthe"
_isEmployed = False

```

```python
## case sensitivity

name = "Shravan"
Name = "Krish"
name == Name
```

```python
## Understanding variable types
## Python is dynamically typed: type of the variable is determined at the runtime
salary = 20000
weight = 50.4
name = "shravan"
is_Indian = True

print(type(salary))
print(type(weight))
print(type(name))
print(type(is_Indian))
```

```python
## Type conversion
age = 25
print(type(age))
# we need to convert this age which is int to string
# so we can use the str function
age_str = str(age)
print(age_str)
print(type(age_str))
```

```python
# string to int
salary = "25000"
print(type(salary))
# converting the string type to integer using the int function
int_salary = int(salary)
print(type(int_salary))
```

```python
name = "Shravan"
print(int(name)) # invalid literal
```

```python
temperature = 25.6
print(type(temperature))
temperature_in_int = int(temperature) ## converting float to int
print(temperature_in_int)
print(type(temperature_in_int))
print(float(temperature_in_int)) # converting int to float
```

```python
## Dynamic typing
var = 10
print(var , type(var))

# python allows the type of variable to change as the program executes
var = "ronaldo"
print(var , type(var))
```

```python
## input from the user
age = (input("Enter your age : "))
print(age , type(age)) # the type of age will be string which is by default

# and if we want to change the type of the input we are taking from the user we can use the int function
age = int(input("Enter your age : "))
print(age , type(age))
```

```python
# simple calculator

num1 = float(input("Enter the first  number "))
num2 = float(input("Enter the second number"))

addition = num1+num2
difference = num1-num2
multiplication = num1*num2
division = num1/num2

print("Addition" , addition)
print("Multiplication ",multiplication)
print("Division",division)
print("Difference",difference)
```
