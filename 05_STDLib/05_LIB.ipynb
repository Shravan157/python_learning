```python
## data serialization
# eg: converting dataframes , converting dictionary to json

import json
data = {'name' : 'Shravan' , 'Age' : 20 , 'Course' : 'Python' }
json_str = json.dumps(data)  # we are converting the dictionary to json string
print(json_str)
print(type(json_str))

parsed_data = json.loads(json_str)
print(parsed_data)
print(type(parsed_data))
```

```python
## csv


import csv

with open('example.csv' , mode = 'w' , newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['name' , 'course'])
    writer.writerow(['shravan' , 'Python and Java'])

with open('example.csv' , mode = 'r') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        print(row)
```

```python
## Date time library

from datetime import datetime,timedelta

# using datetime.now() method we can access the current time
now = datetime.now()
print(now)

# using now - timedelta(days = ?) and by mentioning the days in past we can get the time and date
yesterday = now - timedelta(days=1)
print(yesterday)
```

```python
## time module

import time
print(time.time())
time.sleep(2)
print(time.time())
```

```python
## Regular expression
import re
pattern = r'\d+' # This pattern is about digits
text = 'There are 123 apples 456' # we have a normal text
match = re.search(pattern,text) # we are using the search method in re to check the digits against the normal text
print(match.group()) # so digits in the text are 123 which are matching with the pattern so it is returned
```
