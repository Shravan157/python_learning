```python
## Importing modules in python: Modules and packages
## In python modules and packages help organize and reuse code. Here 's a comprehensive guid on how to import them

```

```python
import math
math.sqrt(16)
```

```python
from math import sqrt,pi # how to import specific functions from the library
print(sqrt(16))
print(sqrt(25))
print(pi)
```

```python
import numpy as np
np.array([1,2,3,4])
```

```python
## from math we want to install every module
from math import *
print(sqrt(16))
print(pi)
```

```python
# fist way to import custom packages inside our code
from package.maths import addition
print(addition(2,3))
```

```python
# second way to import custom pacakges inside our code
from package import maths
maths.addition(2,3)
```
