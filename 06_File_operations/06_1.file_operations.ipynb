File operation - Read and write files

File handling is a cruical part of any programming language. python provides built in functions and methods to
read from and write to files, both text and binary.

```python
## Read a file

with open('example.txt' , 'r') as file:
    content = file.read() # reads the entire content of the file
    print(content)
```

```python
## Read a file line by line

with open('example.txt' , 'r') as file:
    for line in file:
        print(line.strip()) # strip() removes the new line character
```

```python
## writing a file(overwriting): when we open the file it considers the file as empty and write the content in it

with open('example.txt' , 'w') as file:
    file.write("write this statement \n")
    file.write("Add this as well \n")
```

```python
## write a file without overwriting, just append instead of write mode

with open('example.txt', 'a') as file:
    file.write("Append operation taking place \n")
    file.write("The document will not be overwritten")
```

```python
##  write multiple lines of text or list of lines
##  using of writelines() function

lines = ['First line\n' , 'second line \n' , 'third line \n' , 'fourth line \n']
with open('example.txt' , 'a') as file:
    file.writelines(lines)
```

Working with Binary Files


```python
## write to a binary file

data = b'\x00\x01\x02\x03\x04' # This is sample data in binary format

with open('example.bin' , 'wb') as file: # mode is write byte and read byte
    file.write(data)
```

```python
# reading from the bin file
with open('example.bin' , 'rb') as file:
    print(file.read())
```

```python
### Read the content from a source text file and write it to a destination text file

with open('demo.txt' , 'r') as src_file:
    content = src_file.read()

with open('destination.txt' , 'w') as destination_file:
    destination_file.write(content)
```

```python
## similar code but more optimized
def count_text_file(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        line_count = len(lines)
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
    return line_count, word_count, char_count

file_path = 'demo.txt'
lines, words, characters = count_text_file(file_path)
print(f"Lines: {lines}, Words: {words}, Characters: {characters}")
```
