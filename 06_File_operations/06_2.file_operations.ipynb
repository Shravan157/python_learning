```python
## creating a new directory
import os
new_directory = 'package_folder'
os.makedirs(new_directory)
```

```python
## list all the directories and files from a folder
items = os.listdir('.') ## we are extracting all the files and folders between the current directory using '.'
for item in items:
    print(item)
```

```python
### Joining paths : most imp feature of using os module

dir_name = 'folder'
file_name = 'file.txt'
full_path = os.path.join(os.getcwd() ,dir_name, file_name)
print(full_path)
```

```python
path = 'example1.txt'

if os.path.exists(path):
    print(f"The {path} already exists")
else:
    print(f"The {path} does not exist")
```

```python
## checking if a path is a file or directory
import os
path = 'example.txt'
if os.path.isfile(path):
    print(f"The {path} is a file")
elif os.path.isdir(path):
    print(f"The {path} is a directory")
else:
    print(f"The {path} is neither a file nor a directory")
```

```python
## getting the absolute path
relative_path ='example.txt'
absolute_path = os.path.abspath('example.txt')
print(relative_path)
print(absolute_path)
```
