### Assignment 2: Writing to a File

Write a function that writes a list of strings to a file named `output.txt`, with each string on a new line.

```python
# writing a file overwrites the already present text or content within the file
lst_str = ["First text \n" , "second text \n" , "third text \n"]
def file_write(file_path):
    with open(file_path, 'w') as f:
        return f.writelines(lst_str)
file_write('assignment.txt')
```


### Assignment 5: Counting Words in a File

Write a function that reads the contents of a file named `document.txt` and returns the number of words in the file.

```python
def return_words(file_path):
    with open(file_path) as file:
        word_count = 0
        for line in file:
            words = line.split()
            word_count += len(words)
    return word_count
return_words('example.txt')
```

### Assignment 6: Finding and Replacing Text

Write a function that finds and replaces all occurrences of a given word in a file named `data.txt` with another word.

```python
def find_and_replace(file_path , new_word , old_word ):
    with open(file_path , 'r') as file:
        content = file.read()
        print(content)
    with open(file_path , 'w') as file:
        new_word = content.replace(old_word, new_word)
        file.write(new_word)
find_and_replace('example.txt' , 'replaced line' , 'First line')
```

### Assignment 7: Reading a File in Reverse

Write a function that reads the contents of a file named `reverse.txt` and prints each line in reverse order.

```python
def rev_line(file_path):
    with open(file_path, 'r') as file:
        line = file.readlines()

    for line in reversed(line):
        print(line)
rev_line('example.txt')
```


### Assignment 8: Counting Lines, Words, and Characters

Write a function that reads the contents of a file named `stats.txt` and returns the number of lines, words, and characters in the file.

```python
def function_read(file_path):
    with open(file_path) as f:
        word_count = 0
        char_count = 0
        line_count = 0
        for line in f:
            line_count += 1
            char_count += len(line)
            word_count += len(line.split())
    return word_count, char_count, line_count
function_read('example.txt')
```

### Assignment 9: Merging Multiple Files

Write a function that merges the contents of multiple files into a single file named `merged.txt`.

```python
def merge_files(file_list, output_file):
    with open(output_file , 'w') as outfile:
        for file_name in file_list:
            with open(file_name , 'r') as infile:
                outfile.write(infile.read() + '\n')

merge_files(['example.txt', 'demo.txt', 'assignment.txt'], 'merged.txt')

## code explain: we are taking a list of files and are merging the content into one single file
## we are taking two parameters one is file_list and second is the output_file or the mergedfile
## we are opening the output_file in write mode to enter the data in it
## then we are iterating through the file_list to access each file inorder to modify its content
## we are opening each file in read mode and we are writing the read content in the output_file
```

Assignment 10: Splitting a Large File
Write a function that splits a large file named large.txt into smaller files of 100 lines each.

```python
def split_file(filename , lines_per_file):
    with open(filename , 'r') as file:
        lines = file.readlines()
        file_part_number = 1
        for i in range(0,len(lines),lines_per_file):
            file_part = f'{filename}_part{file_part_number}'
            with open(file_part , 'w') as update_file_part:
                current_line = lines[i:i+lines_per_file]
                for line in current_line:
                    update_file_part.writelines(line)
            file_part_number +=1
split_file('example.txt' , 50)


# The function takes a filename and the number of lines each part file should contain.
#
# It opens the main file and reads all lines into a list using readlines().
#
# A loop runs from 0 to the total number of lines, jumping by lines_per_file each time.
#
# For each loop, a new part file name  is created.
#
# That part file is opened in write mode.
#
# Using slicing (lines[i : i + lines_per_file]), a chunk of lines is selected and written into the part file.
#
# The part number increases for every new file created.
```

Assignment 11: Creating a Log File
Write a function that creates a log file named activity.log and writes log messages with timestamps.


```python
from datetime import datetime
def log_file(filename , message):
    with open(filename , 'a') as logfile:
        logfile.write(f'{message} made at {datetime.now()} \n')
log_file('example.log' , 'Log1')
log_file('example.log' , 'Log2')
```

Assignment 12: Binary File Operations
Write a function that reads a binary file named image.bin and writes its contents to another binary file named copy_image.bin.

```python
def bin_operation(source, destination): # read binary mode
    with open(source , 'rb') as src:
         content = src.read()
         with open(destination , 'wb') as des:
             des.write(content)
bin_operation('source.bin' , 'destination.bin')
```

Assignment 13: CSV File Operations
Write a function that reads a CSV file named data.csv and prints its contents as a list of dictionaries.

```python
# using inbuilt DictReader method
import csv

def read_csv_to_dict(filename):
    with open(filename, 'r') as file:
        reader = csv.DictReader(file)
        return list(reader)

read_csv_to_dict('sample.csv')
```

Assignment 14: JSON File Operations
Write a function that reads a JSON file named data.json and prints its contents as a Python dictionary.

```python
import json
def json_to_dict(file_name):
    with open(file_name , 'r') as json_file:
        content = json.load(json_file)
        return content
json_to_dict('sample.json')
```

Assignment 15: File Permission Handling
Write a function that attempts to read a file named protected.txt and handles any permission errors gracefully by printing an error message.

```python
def read_file(filename):
    try:
        with open(filename , 'r') as file:
            content = file.read()
            print(content)
    except PermissionError as e:
        print(f"Permission denied {e}")
```
