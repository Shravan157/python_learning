```python
## simple example of function

def is_even_or_odd(num):
    """ Whether the given number is even or odd. """
    if num>0:
        if num%2==0:
            print(f"{num} is Even")
        else:
            print(f"{num} is Odd")
    else:
        print(f"{num} is Negative")



is_even_or_odd(5)
is_even_or_odd(6)
is_even_or_odd(7)
```

```python
## working with multiple parameters

def add(num1 , num2):
    return num1+num2

result = add(2,5)
print(result)
```

```python
def greet(name):
    print(f"Hello {name} welcome to function")

greet("shravan")
```

```python
# In case if the user does not pass the value in function, then  we can set a default value
# If the no parameter is entered then the default value will be executed
def greet(name = "Guest"):
    print(f"Hello {name} welcome to function")

greet()
```

```python
## Variable Length arguments
## Positional and keywords arguments
## If there are multiple arguments to be entered in the function, then instead of mentioning all of them we can simply use
## *args
def print_numbers(*args):
    """ Positional arguments : use *args , which has multiple arguments """
    for i in args:
        print(i)

print_numbers(1,2,3,"shravan",5,6) # we can add as many argument as we need
```

```python
## Keywords argument: it is denoted by ** , and all the parameters will be in form of key value pairs

def print_detail(**args):
    for key,value in args.items():
        print(f"{key}:{value}")

print_detail(Name = "Shravan" , City = "Mumbai" , Age = 20 , Course = "CSE(AIML)")
```

```python
## Combining both the Positional Argument and keyword argument in a single function
def print_detail(*args ,**kargs):
    for val in args:
        print(f"Positional Argument : {val}")
    for key,value in kargs.items():
        print(f"{key}:{value}")

print_detail(1,2,4,5,6,Name = "Shravan" , relationship_status = "single" , Age = 20 , Course = "CSE(AIML)" )
```
