Lambda function are small anonymous functions defined using the lambda keyword. They can have any number of arguments
but only  one expression. They are commonly used for short operations or as arguments to higher order functions

```python
## Syntax : lambda which is the name , arguments like the parameters to be entered and the expression which describes the  operation to be performed
lambda arguments: expression
```

```python
# Normal function
def addition(a,b):
    return a+b
```

```python
addition = lambda a,b: a+b # example of lambda function
print(type(addition))
print(addition(3,4))
```

```python
# Normal function to check if the function is even
def even(n):
    if n % 2 == 0:
        return True
    else:
        return False
even(25)
```

```python
# similar function using lambda
is_even = lambda n:n%2==0
print(is_even(25))
```

```python
## Example with multiple parameters

def addition(x,y,z):
    return x+y+z

print(addition(1,2,3))

## How to convert above function to lambda function
return_sum = lambda x,y,z: x+y+z
print(return_sum(1,5,3))
```

```python
## map function

numbers = [1,2,3,4,5,6]
# Instead of using looping we can use map function along with lambda function
# we use map() and pass first argument as the lambda function and the second parameter as iterables
# map() uses lazy loading technique
list(map(lambda x:x**2,numbers))
```
