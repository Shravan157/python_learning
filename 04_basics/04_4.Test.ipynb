````python
students = [
    {'name': 'Rahul', 'marks': [85, 90, 78, 92, 88]},
    {'name': 'Priya', 'marks': [76, 82, 91, 85, 89]},
    {'name': 'Arjun', 'marks': [92, 95, 88, 90, 93]},
    {'name': 'Sneha', 'marks': [65, 70, 68, 72, 71]},
    {'name': 'Vikram', 'marks': [58, 62, 55, 60, 59]}
]


### Tasks to Complete:

# **1. Calculate Average Marks (using `map()` with named function)**
# - Create a function `calculate_average(student)` that returns the average of marks
# - Use `map()` to get a list of all averages

def calculate_average(student):
    marks = student['marks']
    return sum(marks)/len(marks)

avg_student_marks = (list(map(calculate_average,students)))
print(avg_student_marks)

# **2. Add Grade Letter (using `map()` with `lambda`)**
# - Add a 'grade' key to each student based on average:
#   - A: >= 90
#   - B: >= 80
#   - C: >= 70
#   - D: >= 60
#   - F: < 60
# - Return a list of dictionaries with the grade added
add_grades = list(map(
    lambda student: {
        **student,
        'grade': (
            'A' if calculate_average(student) >= 90 else
            'B' if calculate_average(student) >= 80 else
            'C' if calculate_average(student) >= 70 else
            'D' if calculate_average(student) >= 60 else
            'F'
        )
    },
    students
))

print("Grades:", add_grades)


# Use map to apply this to all students


#
# **3. Get Pass/Fail Status (using `map()` with `lambda`)**
# - Create a list of tuples: `(name, "Pass"/"Fail")`
# - Pass criteria: average >= 60
#
# **4. Bonus Challenge: Apply Curve**
# - Add 5 marks to each individual mark (use nested `map()`)
# - Recalculate averages with the curved marks
#
# ### Expected Output Format:
# ```
# Averages: [86.6, 84.6, 91.6, 69.2, 58.8]
# Grades: [{'name': 'Rahul', 'grade': 'B', ...}, ...]
# Pass/Fail: [('Rahul', 'Pass'), ('Priya', 'Pass'), ...]
````
