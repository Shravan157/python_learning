```python
# TASK 10: Combining *args and **kwargs
# Create a function called 'print_all_info' that:
# - Accepts any number of positional arguments (*args)
# - Accepts any number of keyword arguments (**kwargs)
# - Prints "Positional arguments:" followed by all args
# - Prints "Keyword arguments:" followed by all kwargs key-value pairs
# Test with: print_all_info(1, 2, 3, name="Alice", age=25, city="Mumbai")
# YOUR CODE HERE:

def print_all_info(*args , **kwargs ):
    for i in args:
        print(i)
    for key,val in kwargs.items():
        print(f"{key} : {val}")
print_all_info(1, 2, 3, name="Alice", age=25, city="Mumbai")
```

```python
# TASK 12: Check Prime Number
# Create a function called 'is_prime' that:
# - Takes a number as parameter
# - Returns True if the number is prime, False otherwise
# - A prime number is only divisible by 1 and itself
# Test with: 7, 10, 13, 20
# YOUR CODE HERE:
def is_prime(num):
    if num>1:
        for i in range(2,num):
            if num%i == 0:
               return False
    else:
        print("please enter a valid number")
    return True
print(is_prime(7))
print(is_prime(10))
print(is_prime(13))
print(is_prime(20))
```

```python
# TASK 14: List Statistics
# Create a function called 'calculate_statistics' that:
# - Takes any number of numeric arguments using *args
# - Returns three values: average, minimum, maximum
# Test with: 45, 67, 23, 89, 12, 56, 78
# YOUR CODE HERE:

def calculate_statistics(*args):
    min_value = min(args)
    max_value = max(args)
    avg_value = sum(args)/len(args)
    return min_value,max_value,round(avg_value,2)

calculate_statistics(45, 67, 23, 89, 12, 56, 78)
```

```python
# TASK 15: Student Grade Calculator
# Create a function called 'calculate_grade' that:
# - Takes a score (0-100) as parameter
# - Returns a grade based on:
#   90-100: 'A'
#   80-89: 'B'
#   70-79: 'C'
#   60-69: 'D'
#   Below 60: 'F'
# Test with: 95, 82, 76, 65, 45
# YOUR CODE HERE:

def calculate_grade(score):
    if 0<score<100:
        if 90<score<100:
            grade = "A"
        elif 80<score<89:
            grade = "B"
        elif 70<score<89:
            grade = "C"
        elif 60<score<69:
            grade = "D"
        else:
            grade = "F"
        return grade
    else:
        return "Enter a valid value between (0-100)"

print(calculate_grade(95))
print(calculate_grade(82))
print(calculate_grade(76))
print(calculate_grade(65))
print(calculate_grade(45))

```

```python
# TASK 16: Challenge - Shopping Cart
# Create a function called 'calculate_total' that:
# - Accepts item prices using *args
# - Accepts tax_rate and discount as **kwargs (optional)
# - Default tax_rate = 0.08 (8%), default discount = 0

# - Calculate: subtotal = sum of all prices
# - Apply discount: discounted_price = subtotal - (subtotal * discount / 100)
# - Apply tax: final_total = discounted_price + (discounted_price * tax_rate)
# - Return the final total
# Test with: calculate_total(100, 200, 150, tax_rate=0.1, discount=15)
# YOUR CODE HERE:


def calculate_total(*args , **kwargs):
    tax_rate = kwargs.get("tax_rate" , 0.08)
    discount = kwargs.get("discount" , 0)

    subtotal = sum(args)
    discounted_price = subtotal - (subtotal * discount / 100)
    final_total = discounted_price + (discounted_price * tax_rate)
    return final_total
calculate_total(100, 200, 150, tax_rate=0.1, discount=15)
```

```python
# TASK 17: Challenge - Text Analyzer
# Create a function called 'analyze_text' that:
# - Takes a text string as parameter
# - Returns a tuple with:
#   1. Number of words
#   2. Number of characters (excluding spaces)
#   3. Number of vowels
# Test with: "Python programming is fun and powerful"
# YOUR CODE HERE:

def analyze_text(text):
    word_count = len(text.split())
    char_count = 0
    for i in text:
        if i !=" ":
            char_count = char_count+1

    vowel_count = 0
    for vowels in text:
        if vowels in "aeiouAEIOU":
            vowel_count = vowel_count+1

    return (word_count , char_count , vowel_count)

print(analyze_text("Python programming is fun and powerful"))
```

```python
# TASK 18: Challenge - Number Pattern Generator
# Create a function called 'generate_pattern' that:
# - Takes a number 'n' as parameter
# - Takes pattern_type as optional parameter (default: "numbers")
# - If pattern_type is "numbers": print numbers from 1 to n, each on new line
# - If pattern_type is "stars": print asterisks from 1 to n (pyramid pattern)
# Test with:
# 1. generate_pattern(5)
# 2. generate_pattern(5, "stars")
# YOUR CODE HERE:

def generate_pattern(n , pattern_type = 'numbers'):
    if pattern_type == 'numbers':
        for i in range(1,n+1):
            print(f"{i}\n")
    if pattern_type == 'star':
        for j in range(1,n+1):
            for k in range(1,j+1):
                print("*" , end="")
            print()
generate_pattern(5 , 'star')
```

```python
# TASK 19: Challenge - Dictionary Builder
# Create a function called 'build_person_dict' that:
# - Takes 'name' and 'age' as required parameters
# - Accepts any additional information using **kwargs
# - Returns a dictionary containing all the information
# Test with: build_person_dict("Alice", 25, city="Mumbai", profession="Engineer", hobby="Reading")
# YOUR CODE HERE:

def build_person_dict(name, age, **kwargs):
    person_details = {
        "name": name,
        "age": age
    }
    for key, value in kwargs.items():
        person_details[key] = value

    return person_details


print(build_person_dict("Alice", 25, city="Mumbai", profession="Engineer", hobby="Reading"))
build_person_dict("Alice", 25, city="Mumbai", profession="Engineer", hobby="Reading")
```
