The map() function in python

The map() function applies a given function to all items in an input list(or any other iterable) and returns
a map object. This is particularly useful for transforming data in a list comprehensively

```python
## example and syntax

def square(x):
    return x**2

numbers = [1,2,3,4,5,6,7,8]
print(list(map(square,numbers)))
```

```python
## Lambda function with map
num = [1,2,3,4,5]
print(list(map(lambda x:x**2,num)))
```

```python
## Mapping of multiple iterables

num1 = [1,2,3]
num2 = [4,5,6]

added_num = list(map(lambda x,y : x+y , num1,num2 ))
print(added_num)
```

```python
## map() to convert list of string to integers
str_num = ['1','2','3','4','5']
# we have to convert the above string into integers using map function

int_num = list(map(int,str_num)) # why int , because we are type_casting the string to int
print(int_num)
```

```python
## Inbuilt function in map

words = ['apple' , 'banana' , 'cherry']
upper_words = list(map(str.upper,words))
print(upper_words)
```

```python
# consider there is a dictionary which is person , which is taken as paramter in function
# and we are returning the value of the 'name' key in the person dictionary
def get_name(person):
    return person['name']

people = [
    {'name' : 'Shravan' , 'age' : 20 , 'course' : 'Java'},
    {'name' : 'Ayush' , 'age' : 20 , 'course' : 'Python'}
]

list_names = list(map(get_name,people))
print(list_names)
```
