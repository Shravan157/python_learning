#### Iterators in python 
Iterators are advanced python concepts that allow for efficient looping and memory management. Iterators provide
a way to access elements of a collection sequentially without exposing the underlying structure.An iterator does not load all elements at once.
It keeps track of the current position and returns the next value only when next() is called.

```python
my_list = [1,2,3,4,5,6]

## if i have to iterate through the list
for elements in my_list:
    print(elements)
print(type(my_list))
print(my_list)
```

```python
## How to create an iterator
my_iterator = iter(my_list)
print(type(my_iterator)) ## type <class 'list_iterator'>
print(my_iterator) # we get only the memory location, it uses the lazy loading technique
## we will not get the element unless and until we iterate through it 
```

```python
## how to iterate through all the elements
## we make use of a function called as next() and we pass our iterator as an argument in it 
## it gives the first element of the list
next(my_iterator) ## access only one value at a time
```

```python
list_num = [2,3,4,5,6,7,8,9]
my_iterator = iter(list_num)

try:
    print(next(my_iterator))
    print(next(my_iterator))  # 2
    print(next(my_iterator))  # 3
    print(next(my_iterator))  # 4
    print(next(my_iterator))  # 5
    print(next(my_iterator))  # 6
    print(next(my_iterator))  # 7
    print(next(my_iterator))  # 8
    print(next(my_iterator))  # 9
    print(next(my_iterator))  # StopIteration exception raised!
except StopIteration:
    print("No more elements in the elements")
```

```python
## String iteration
text = "Hello"
text_iter = iter(text)

print(next(text_iter))  # H
print(next(text_iter))  # e
print(next(text_iter))  # l
print(next(text_iter))  # l
```

```python
my_dict = {'a': 1, 'b': 2, 'c': 3}

# Iterate over keys
key_iter = iter(my_dict.keys())
print(next(key_iter))  # a

# Iterate over values
value_iter = iter(my_dict.values())
print(next(value_iter))  # 1

# Iterate over items
item_iter = iter(my_dict.items())
print(next(item_iter))  # ('a', 1)
```

```python
# Files are iterators by default
with open('file.txt') as f:
    print(next(f))  # First line
    print(next(f))  # Second line
```

```python
range_iter = iter(range(5))
print(next(range_iter))  # 0
print(next(range_iter))  # 1
```

Why Iterators Are Used
1. Memory Efficiency:
Iterators generate values on-the-fly (lazy evaluation) rather than storing all values in memory at once.
2. Work with Infinite Sequences:
3. Pipeline Processing: You can chain iterators for efficient data processing:

```python
# This creates a list in memory (takes up space)
numbers_list = [x**2 for x in range(1000000)]

# This creates an iterator (minimal memory)
numbers_iter = (x**2 for x in range(1000000))
```

```python
def infinite_counter():
    num = 0
    while True:
        yield num
        num += 1

counter = infinite_counter()
print(next(counter))  # 0
print(next(counter))  # 1
# Can continue indefinitely
```

```python
numbers = range(1000000)
squares = (x**2 for x in numbers)
even_squares = (x for x in squares if x % 2 == 0)
# Nothing computed yet until you iterate
```
