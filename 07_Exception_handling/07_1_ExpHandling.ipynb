#### Understanding Exceptions
Exception handling in python allows you to handle errors gracefully and take corrective actions
without stopping the execution of program. How to use try,except,else and finally blocks


### What are exceptions?
Exceptions are events that disrupt the normal flow of a program. They occour when an error is
encountered during program execution. common exceptions include:
1.ZeroDivisionError: Dividing by zero
2.FileNotFoundError: File not found
3.ValueError: Invalid value
4.TypeError: Invalid type

```python
## Here it b is not defined it is a NameError
## How to handle it, enter the code in the try block which might cause error
## if it causes error do not stop the execution of the program continue to the except block
try:
    a=10
    a=b
except NameError as exc:
    print(f'{exc}')
```

```python
# Example of zero division error
try:
    result = 1/0
except ZeroDivisionError as exc:
    print(f'{exc}')
    print(f'Please enter the denominator greater than zero')
```

```python
## All the exception are derived from a same exception class in python
## But what if we don't know what type of exception occurs
## All the exception are defined from a same class which is Exception
## It is IMP to write the main exception block at the end because it is the parent class
try:
    result = 1/2
    a = b
except ZeroDivisionError as ex:
    print(f'{ex}')
except Exception as ex1:
    print(f'{ex1}')
    print('main exception got caught here')
```

```python
## Example: Handling multiple catch blocks
## catch1: handles that the input made by the user is valid, it must be an integer value
## catch2: handles that the denominator is not zero

try:
    num = int(input('Enter a number: '))
    result = 10/num
    print(result)
except ValueError as exc:
    print(f'{exc}')
except ZeroDivisionError as exc:
    print(f'Cannot divide by zero: {exc}')
except Exception as exc:
    print(f'{exc}')
```

```python
## try, except and else block

try:
    nums = int(input('Enter a number: '))
    result = 10/nums
except ValueError:
    print('Enter a valid number')
except ZeroDivisionError:
    print('Cannot divide by zero')
except Exception as ex:
    print(f'{ex}')
else:
# If the try block executed and there is no exception then, close the try,except and i want to execute some block of code anyhow
    print('Result =', result)
```

```python
## try , except , else , finally
## try: has all our block of code where exception might arise
## except: based on the input value the catch block will execute
## else: if no exception then else block will execute
## finally: code written will get executed even if the number is entered or not entered (it has to execute)
## does not matter what might happen, finally will get always executed
try:
    num = int(input('Enter a number: '))
    result = 10/num
except ValueError:
    print('Enter a valid number')
except ZeroDivisionError:
    print('Cannot divide by zero')
else:
    print('Result =', result)
finally:
    print('Execution complete')
```

```python
## Example1: File and exception handling
## try: contains the code which might cause exception
## except: handles the exception in the try block
## finally: as we have opened the file, closing of the file must be performed after the execution of the program
try:
    with open('sample1.txt' , 'r') as file:
        content = file.read()
        print(content)
except Exception as exc:
    print(f'{exc}')
finally:
    if 'file' in locals() or not file.close():
        file.close()
        print('File closed')
```
