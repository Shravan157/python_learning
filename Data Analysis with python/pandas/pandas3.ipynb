**Pandas Dataframe**
A pandas DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns)

*Creating DataFrame*
1. using list
2. using dictionaries

**Filtering a Dataframe**

```python
# find all final winners
mask = ipl_data['MatchNumber'] == 'Final'
final_match_df = ipl_data[mask]
final_match_df[['Season','WinningTeam']]

## one liner
ipl_data[ipl_data['MatchNumber'] == 'Final'][['Season' , 'WinningTeam']]
```

```python
# how many super overs finishes have occured
ipl_data[ipl_data['SuperOver'] == 'Y'].shape[0]
```

```python
## how many matches has csk won in kolkata
csk_won_kolkata = (ipl_data[ipl_data['City'] == 'Kolkata']['WinningTeam'] == 'Chennai Super Kings').sum()
print(f'Csk Wins in Kolkata {csk_won_kolkata}')


## alternative
ipl_data[(ipl_data['City'] == 'Kolkata') & (ipl_data['WinningTeam'] == 'Chennai Super Kings')].shape
```

```python
# toss winner is match winner in percentage
(ipl_data[ipl_data['TossWinner'] == ipl_data['WinningTeam']].shape[0]/ipl_data.shape[0]) * 100
```

```python
# movies with rating higher than 8 and votes > 10000
movies[(movies['imdb_rating'] > 8) & (movies['imdb_votes'] > 10000)].shape[0]
```

```python
# Action movies with rating higher than 7.5
mask1 = movies['genres'].str.split('|').apply(lambda x:'Action' in x)
# mask1 = movies['genres'].str.contains('Action')
mask2 = movies['imdb_rating'] > 7.5
movies[mask1 & mask2]
```

```python
# write a function that can return the track record of 2 teams against each other

def track_record(team1 , team2):
  team1_vs_team2 = ipl_data[((ipl_data['Team1'] == team1) & (ipl_data['Team2'] == team2)) | ((ipl_data['Team1'] == team2) & (ipl_data['Team2'] == team1))]

  team1_wins =(team1_vs_team2['WinningTeam'] == team1).sum()
  team2_wins =(team1_vs_team2['WinningTeam'] == team2).sum()

  return team1_wins , team2_wins

team1_w , team2_w = track_record('Royal Challengers Bangalore' , 'Chennai Super Kings')
print(f'RCB wins {team1_w}')
print(f'CSK wins {team2_w}')
```
