**Slicing Indexing on pandas series**

There are two ways to access data
1. iloc(integer location based indexing)
2. loc(location based indexing)

```python
s = pd.Series([10,20,30,40,50,60] , index = ['a','b','c','d','e','f'])
print(s)
```

```python
## accessing values using iloc
print(s.iloc[1]) #20
print(s.iloc[-1]) # 60

## accessing values using loc
print(s.loc['d'])
```

```python
## Negative indexing does not work in loc
random_series = pd.Series([1,2,3,4,5], index = [1,2,3,4,5])
print(random_series)
```

```python
print(random_series.loc[5])
```

```python
import pandas as pd
import numpy as np

# Series 1: Integer index labels
s1 = pd.Series([100, 200, 300, 400, 500], index=[5, 3, 1, 8, 2])

# Series 2: String index labels
s2 = pd.Series([10, 20, 30, 40, 50], index=['b', 'd', 'a', 'e', 'c'])

# Series 3: Non-sequential integer labels
s3 = pd.Series(['A', 'B', 'C', 'D', 'E'], index=[10, 20, 30, 40, 50])

# Series 4: Mixed type index (for advanced confusion)
s4 = pd.Series([1.1, 2.2, 3.3, 4.4], index=[0, '1', 2, '3'])
```

```python
# What does this return?
result = s1.iloc[2]
print(result)
# A) 300  B) 200  C) Error  D) 100

# What does this return?
result = s1.loc[2]
print(result)
# A) 500  B) 300  C) Error  D) 200

# What does this return?
result = s1.loc[1]
print(result)
# A) 300  B) 200  C) 100  D) Error

# What does this return?
result = s2.loc['a':'c']
print(result)
# A) ['a', 'b', 'c'] values  B) [30, 10, 50]  C) Error  D) [30, 50]

# What does this return?
result = s2.iloc[1:3]
print(result)
# A) ['d', 'a']  B) [20, 30]  C) ['a', 'e']  D) Error
```
