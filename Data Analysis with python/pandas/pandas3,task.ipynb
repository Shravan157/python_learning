###`Q-8:` Team1 vs Team2 Dashbord
Create a function which will take two string(name of two teams) as input. Show win Loss record between them and player getting most player of the match award in matches between these two teams.
```
team1_vs_team2('Kolkata Knight Riders','Chennai Super Kings')
```

````python
def track_record(team1,team2):
  team1_vs_team2 = ipl[((ipl['Team1'] == team1) & (ipl['Team2'] == team2)) | ((ipl['Team1'] == team2) & (ipl['Team2'] == team1))]
  team1_w = (team1_vs_team2['WinningTeam'] == team1).sum()
  team2_w = (team1_vs_team2['WinningTeam'] == team2).sum()
  player_of_match = team1_vs_team2['Player_of_Match'].value_counts(ascending = False).head(1)
  return team1_w,team2_w,player_of_match , team1_vs_team2
team1_wins,team2_wins,most_pom_wins, Head_to_Head = track_record('Chennai Super Kings','Mumbai Indians')
print(f'Total Head to Head Matches {Head_to_Head.shape[0]}')
print(f'Chennai Super kings wins = {team1_wins}')
print(f'Mumbai Indians wins = {team2_wins}')
print(f'Most POM Awards {most_pom_wins}')
````
