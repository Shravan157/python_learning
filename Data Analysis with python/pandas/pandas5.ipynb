9. *isnull()* (series + dataframe)

```python
students = pd.DataFrame({
    'Name' : ['nitish','ankit','rupesh',np.nan,'shravan',np.nan,'akshay',np.nan,'vedant',np.nan],
    'College': ['bit','IIT','VIT',np.nan,np.nan,'vjti','ssit',np.nan,np.nan,'KJ'],
    'Branch' : ['eee','it','cse',np.nan,'me','ce','civ','cse','bio',np.nan],
    'cgpa' : [6.66,8.25,6.41,np.nan,5.60,9.00,7.40,10.00,7.40,np.nan],
    'package': [4.0,5.0,6.0,np.nan,6.0,7.0,8.0,9.0,np.nan,np.nan]
})
students
```

```python
students['Name'].isnull() # checks each value if it is null , if yes then true else false
```

10. *notnull()* (series + dataframe)

opposite of the isnull() method

```python
students['Name'].notnull() # if not null then true , if null then false
```

10. *hasnans* (series) -> Attribute

Question -> Is there null value in the mentioned dataframe column
hasnans -> If present -> True  | If not -> False

```python
students['Name'].hasnans
```

```python
students.isnull()
students.notnull()
```

11. *dropna()* (series+dataframe)


Disadvantage -> Removes data

Default Behaviour is how = 'any'


```python
students['Name'].dropna() # on a series
students.dropna(how = 'any') # removes the rows which contains even single null values
```

```python
students.dropna(how = 'all') # until and unless all the rows have missing values it should not drop the null values
```

```python
## we can delete or drop the null values based on a specific column
## Eg : Drop the null values only if the name column has any null value
students.dropna(subset='Name')
```

```python
## we can do it based on multiple columns as well
students.dropna(subset=['Name','College'])
```
