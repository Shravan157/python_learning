12. fillna() (series + dataframe)

Filling missing values

```python
# Filling null values in the Name column with a string called as 'Unknown'
students['Name'].fillna('Unknown').hasnans
```

```python
# using fill na on  dataframe
students.fillna(0) # filling all missing values with zero
students['package'].fillna(students['package'].mean())
```

*ffill() and bfill()*

Forward fill and backward fill




```python
students['Name'].ffill() # forward fill , fill the missing value with the value forward to it
students['College'].bfill() # backward fill
```

14. *apply()*

To apply custom logic on a dataframe or series

```python
## Calculate the sigmoid of each value in the series
temp = pd.Series([10,20,30,40,50])

def sigmoid(values):
  return 1/1+np.exp(-values)

temp.apply(sigmoid)
```

```python
## example: create a new column eucladian distance of the both point
points_df = pd.DataFrame({
    'Point1' : [(3,4),(-6,5),(0,0),(-10,1),(4,5)],
    'Point2' : [(-3,4),(0,0),(2,2),(10,10),(1,1)]
})

points_df
```

```python
def eucladian(row):
  pt_a = row['Point1']
  pt_b = row['Point2']

  return ((pt_a[0] - pt_b[0])**2 + (pt_a[1] - pt_b[1])**2)**0.5
```

```python
points_df['Eucladian Distance'] = points_df.apply(eucladian , axis = 1)
points_df
```
