##### Series Indexing & Slicing

In Pandas, we have two primary ways to access data by index:

1. **`iloc` (Integer-location based indexing)**: 
   - Uses 0-based integer positions (like a Python list).
   - Always uses integers (0 to length-1).
   - Slicing: `[start:stop]` is **exclusive** of the stop index.

2. **`loc` (Label-based indexing)**:
   - Uses the **Labels** of the index.
   - Labels can be strings, integers, or any type defined in the index.
   - Slicing: `[start:stop]` is **inclusive** of both start and stop indexes.

````python
# Example 1: Series with String Index
s = pd.Series([10, 20, 30, 40, 50], index=['a', 'b', 'c', 'd', 'e'])

print("Series s:")
print(s)

print("\n--- Accessing single elements ---")
print("s.iloc[0] (Positional 0):", s.iloc[0])
print("s.loc['a'] (Label 'a'):   ", s.loc['a'])

print("\n--- Slicing Differences ---")
print("s.iloc[0:3] (Exclusive of position 3):\n", s.iloc[0:3])
print("s.loc['a':'c'] (Inclusive of 'c'):\n", s.loc['a':'c'])

print("\n--- Negative Indexing (Works with iloc) ---")
print("s.iloc[-1] (Last element):", s.iloc[-1])
````

##### The Confusion: Integer Labels

When the index itself contains integers that are NOT 0, 1, 2..., code like `s[1]` can be ambiguous. It is always better to use `loc` or `iloc` to be explicit.

````python
# Example 2: Series with custom Integer Index
s2 = pd.Series([100, 200, 300], index=[10, 20, 30])

print("Series s2 (Labels: 10, 20, 30):")
print(s2)

print("\n--- loc vs iloc on integer labels ---")
print("s2.iloc[0] (Physical First element):", s2.iloc[0])  # Output: 100
print("s2.loc[10] (Label 10):              ", s2.loc[10]) # Output: 100

# s2.loc[0] would raise an error because label 0 doesn't exist!
# s2.iloc[10] would raise an error because position 10 is out of bounds!
````

##### Boolean Indexing on Series

````python
import pandas as pd

series = pd.Series([10, 20, 30, 40, 50], index=['A', 'B', 'C', 'D', 'E'])

# Step 1: Create a boolean mask
mask = series > 25
print(mask)
# A    False
# B    False
# C     True
# D     True
# E     True

# Step 2: Use the mask to filter
result = series[mask]
print(result)
# C    30
# D    40
# E    50
````
