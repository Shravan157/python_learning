```python

# Remove Duplicate Emails (Using set() to eliminate duplicates)
#
# You receive a list of emails from a form. Some users typed duplicates.
#
# Task
# emails = ["a@gmail.com", "b@gmail.com", "a@gmail.com", "c@gmail.com", "b@gmail.com"]
#
#
# Convert the list to a set
#
# Print the unique emails
#
# Convert back to a list
#
# (Concepts used: creating set, uniqueness)

emails = ["a@gmail.com", "b@gmail.com", "a@gmail.com", "c@gmail.com", "b@gmail.com"]
emails_set = set(emails)
print(emails_set)
email_set = list(emails_set)
print(email_set)

```

```python
# Check If a Product Is Available (Membership Test)
#
# You have a set of products in stock:
#
# products = {"shoes", "tshirt", "watch", "bag"}
#
# Task
#
# Ask the user for a product name and check:
#
# If it exists → print “Available”
#
# If not → print “Out of stock”
#
# (Concepts used: in operator)

products = {"shoes", "tshirt", "watch", "bag"}
user_response = (input("Please enter your product name: "))
print(user_response in products)
```

```python
# Add/Remove Items From a Cart (add, remove, discard)
# cart = {"laptop", "mouse"}
#
# Task
#
# Add "keyboard" using add()
#
# Remove "mouse" using remove()
#
# Try to remove "tablet" using both remove() and discard() and observe the difference
#
# (Concepts used: add, remove, discard)

cart = {"laptop", "mouse"}
cart.remove('mouse')
# cart.remove('tablet')
cart.discard('tablet')
print(cart)
```

```python
# First-Added User Gets Removed (pop method)
#
# A queue system stores usernames:
#
# users = {"ram", "shyam", "geeta", "john"}
#
# Task
#
# Use .pop() and print:
#
# Which user got removed
#
# Remaining set
#
# (Concept: pop is arbitrary / FIFO-like in CPython but unordered)

users = {"ram", "shyam", "geeta", "john"}
users.pop()
print(users)
# ram gets popped out because it was the first element inserted
```

```python
# Clear the Temporary Cache (clear method)
# cache = {"/home", "/login", "/dashboard"}
#
# Task
#
# Clear the cache and print the empty set.

cache = {"/home", "/login", "/dashboard"}
cache.clear()
print(cache)
```

```python
# Combine Followers From Two Users (Union)
# user_A_followers = {1, 2, 3, 4}
# user_B_followers = {3, 4, 5, 6}
#
# Task
#
# Find all unique followers of both users.
#
# (Concept: .union())

user_A_followers = {1, 2, 3, 4}
user_B_followers = {3, 4, 5, 6}
total_followers = user_A_followers.union(user_B_followers)
print(total_followers)
```

```python
# Find Common Purchased Items (Intersection)
# cust1_items = {"milk", "bread", "eggs"}
# cust2_items = {"bread", "paneer", "milk"}
#
# Task
#
# Find which items both customers bought.
#
# (Concept: .intersection())


cust1_items = {"milk", "bread", "eggs"}
cust2_items = {"bread", "paneer", "milk"}
common_items = cust1_items.intersection(cust2_items)
print(common_items)
```

```python
# Update Existing Set With Common Data (Intersection Update)
# s1 = {1, 2, 3, 4}
# s2 = {3, 4, 5, 6}
#
# Task
#
# Apply s1.intersection_update(s2)
# Print the updated s1.


s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}
s1.intersection_update(s2)
print(s1)

```

```python
# Items Present in Warehouse A but Not in Warehouse B (Difference)
# warehouse_A = {"fan", "TV", "AC", "fridge"}
# warehouse_B = {"TV", "AC"}
#
# Task
#
# Find items only A has.
#
# (Concept: .difference())

warehouse_A = {"fan", "TV", "AC", "fridge"}
warehouse_B = {"TV", "AC"}
warehouse_A.difference(warehouse_B)

```

```python
# Remove Items That Also Exist in Another Set (Difference Update)
# A = {1, 2, 3, 4}
# B = {3, 4, 5}
#
# Task
#
# Apply .difference_update(B) to A.

A = {1, 2, 3, 4}
B = {3, 4, 5}

A.difference_update(B)
print(A)
```

```python
# Items Unique to Each App User (Symmetric Difference)
# user1 = {"login", "search", "cart"}
# user2 = {"cart", "payment", "wishlist"}
#
# Task
#
# Find actions that are NOT common.
#
# (Concept: .symmetric_difference())

user1 = {"login", "search", "cart"}
user2 = {"cart", "payment", "wishlist"}
not_common = user1.symmetric_difference(user2)
print(not_common)
```

```python
# Validate Required Permissions (issubset & issuperset)
#
# Required permissions:
#
# required = {"read", "write"}
# user = {"read", "write", "execute"}
#
# Task
#
# Check:
#
# Is required a subset of user?
#
# Is user a superset of required?
#
# (Concept: .issubset(), .issuperset())

required = {"read", "write"}
user = {"read", "write", "execute"}

is_subset = required.issubset(user)
is_superset = user.issuperset(required)

print(is_subset)
print(is_superset)
```

```python
# Unique Words in a Description (Words → Set)
# description = "python is easy and python is powerful"
#
# Task
#
# Split the text
#
# Convert words to set
#
# Print unique words and count
description = "python is easy and python is powerful"
desc_split = description.split() # spilt is used to convert the description to list
print(desc_split)

set_desc = set(desc_split) # we are converting list to set , to remove duplicate items
print(set_desc)
print(len(set_desc))
```
