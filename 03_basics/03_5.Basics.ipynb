#### Real world examples using list in python

List are one of the most commonly used data structures in python


```python
from pygments.styles.dracula import orange

## example 1 : Manage a todo list

todo_list = ["Learn module 2" , "Go for a walk" , "Buy groceries"]

## Adding the task
todo_list.append("schedule th meeting with john")
todo_list.append("Get sleep in the afternoon")

## Remove the task
todo_list.remove("Buy groceries")

## checking if a task is in the list
if "Go for a walk":
    print("Don't forget to go for a walk")

print("Remaining tasks")
print(todo_list)
```

```python
## example 2: organizing student grades

grades = [85,92,78,90,88]

## adding a new grade
grades.append(95)

## calculating average grades
average_grade = sum(grades) / len(grades)
print(average_grade)

## Finding the highest and lowest grades
highest_grades = max(grades)
lowest_grades = min(grades)
print(highest_grades)
print(lowest_grades)
```

```python
## example 3: managing an inventory

inventory = ["apples" , "banana" , "oranges" , "grapes"]

#adding a new item
inventory.append("kiwi")

# removing the item from the stock
inventory.remove("banana")

# checking if an item is out of stock
item = 'oranges'
if item in inventory:
    print(f"{item} available")
else:
    print(f"{item} will be available soon ..")

print(f"{inventory}")
```
