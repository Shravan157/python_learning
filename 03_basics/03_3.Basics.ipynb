```python
"""
Sets : sets are built in datatype in python used to store collections of unique item. They are unordered meaning
that the elements do not follow a specific order, and they do not allow duplicate elements. Sets are useful for
membership test, eliminating duplicate entries , and performing mathematical set operations like union
intersection, difference , and symmetric difference
"""
```

```python
## Creating a set: we use curly braces for creating the set

my_set = {1,2,3,4,5}
print(type(my_set))
print(my_set)
```

```python
# creating an empty set
my_set_empty = set()
print(my_set_empty)
```

```python
# we can have tuples and list inside of set , using typecasting it to set class
my_set_empty = set([1,2,3,4,5,6])
print(my_set_empty)
```

```python
## They store unique items: they ignore the duplicate elements and do not display them
## They are unordered meaning that they do not follow a specific order
set_demo = {1,2,3,4,3,7,8,8}
print(set_demo)
```

```python
## Set operations

## Operation 1 : Adding and removing elements
my_set.add(6) # we can add the element at the end of the set using add method
print(my_set)
```

```python
## Operation 2 : Removing element
my_set.remove(6)
print(my_set)
```

```python
## Operation 3: Discarding an element
# What happens if we try to remove an element which is not present in the set, it will give error
## to avoid this we use discard method
## if the element is not present, it will not throw any error
my_set.discard(11)
print(my_set)
```

```python
## Operation 4: POP method (follows first in first out)
## so whatever was the fist element inserted will be popped first
popped_element = my_set.pop()
print(popped_element)
print(my_set)
```

```python
## Operation 5: clear method (removes all the elements from the set)
my_set.clear()
print(my_set)
```

```python
## Set Membership test
## Whether an element is present in the test or not, or an element is a member of a set or not
## Returns True if the member is present and if not, then False
my_set = {1,2,3,4,5,6}
print(3 in my_set) # we are checking if 3 is present in my_set or not
print(10 in my_set) # we are checking if 10 is present in my_set or not
```

```python
## Mathematical Operation

set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

## Union operation: Combining the elements of both the set without repeating

union_set = set1.union(set2)
print(union_set)


## Intersection operation: Takes only the common elements from the sets
intersection_set = set1.intersection(set2)
print(intersection_set)

## Intersection update: output of the intersection was {something}, that something will get updated in the set1
set1.intersection_update(set2)
print(set1)


## Difference: Return a new set with elements in the set that are not in the others.
set3 = {1,2,3,4,5}
set4 = {4,5,6,7,8}
print(set3.difference(set4))
# | Element | In set4? | Included in result? |
# | ------- | -------- | ------------------- |
# | 1       | No       | ✔ Yes               |
# | 2       | No       | ✔ Yes               |
# | 3       | No       | ✔ Yes               |
# | 4       | Yes      | ❌ No                |
# | 5       | Yes      | ❌ No                |

## Difference update : output of the difference was {something} , which will get updated inn the set
set3.difference_update(set4)
print(set3)
```

```python
## Symmetric difference: unique elements from both the set are combined
##, so the 4,5 were common to both sets, and they are ignored in this case
set5 = {1,2,3,4,5}
set6 = {4,5,6,7,8}
print(set5.symmetric_difference(set6))
```

```python
## Set methods
set_demo1 = {1,2,3}
set_demo2 = {3,4,5}


# issubset() → checks if all elements of set_demo1 exist inside set_demo2
# Here: 1 and 2 are NOT in set_demo2 → so it returns False

print(set_demo1.issubset(set_demo2))
print(set_demo1.issuperset(set_demo2))

# issuperset() → checks if set_demo1 contains all elements of set_demo2
# Here: 4 and 5 are NOT in set_demo1 → so it returns False
```

```python
## Removing duplicates from the list
lst =[1,2,2,3,4,4,5]
set_lst = set(lst)
print(set_lst)
```

```python
## Counting unique words in text
text = "IN this tutorial we are discussing about sets"
words=text.split()

## convert list of words to set to get unique words
unique_words = set(words)
print(unique_words)
print(len(unique_words))
```
