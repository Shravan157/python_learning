```python
# 1. Create a list
#
# Create a list named students with 5 student names and print:
#
# Length of the list
#
# First and last element


student = ["Shravan" , "Sumeet" , "Soham" , "Raj" , "Sahil"]
print(student)
print(len(student))
print(student[0] , student[-1])

```

```python
# 2. Access list elements

# Given:
#
# cities = ["Mumbai", "Delhi", "Pune", "Chennai", "Bangalore"]
#
#
# Print:
#
# The 2nd element
#
# The last two elements using slicing
#
# Elements from index 1 to 3

cities = ["Mumbai", "Delhi", "Pune", "Chennai", "Bangalore"]
print(cities[1]) # as second element is Delhi
print(cities[1:3])
print(cities[3:])
```

```python
# 3. Modify list
#
# Replace "Delhi" with "Hyderabad" in the cities list.

cities[1] = "Hyderabad"
print(cities)
```

```python
# 4. Replace multiple elements using slicing
#
# Given:
#
# nums = [10,20,30,40,50]
#
#
# Replace the last 3 values with:
#
# 100, 200, 300
nums = [10,20,30,40,50]
nums[2:] = [100,200,300]
print(nums)
```

```python
# Insert and append
#
# Create a list of libraries:
#
# libs = ["NumPy", "Pandas", "Matplotlib"]
#
#
# Insert "Seaborn" at index 1
#
# Append "Scikit-learn"

libs = ["NumPy", "Pandas", "Matplotlib"]
libs.insert(1,"Seaborn")
libs.append("Scikit-learn")
print(libs)
```

```python
# 6. Remove items
#
# From the list:
#
# devices = ["laptop", "tablet", "mobile", "laptop", "desktop"]
#
#
# Remove only the first "laptop"
#
# Pop the last item & print it

devices = ["laptop", "tablet", "mobile", "laptop", "desktop"]
devices.remove("laptop")
print(devices)
popped_item = devices.pop()
print(popped_item)
print(devices)

```

```python
# 7. Count occurrences
#
# Count how many times "laptop" appears.
print(devices.count("laptop"))
```

```python
# 8. Sort and reverse
#
# Sort the list alphabetically, then reverse it.

devices.sort()
print(devices)

devices.reverse()
print(devices)
```

```python
# 9. Clear the list
#
# Using .clear(), empty the devices list.
devices.clear()
print(devices)
```

```python
# 10. Slicing task
#
# Given:
#
# nums = list(range(1, 21))
#
#
# Print:
#
# Even-indexed elements
#
# Odd-indexed elements
#
# List in reverse
#
# Every 3rd element

nums = list(range(1, 21))
print(nums[::2]) ## for even
print(nums[1::2]) ## for odd
print(nums[::-1])
print(nums[2::3])
```

```python
# 11. Iterate normally
#
# Create a list of 5 countries; print each using a normal for loop.

lst = ["Netherlands" ,"Brazil" , "France" , "Argentina" , "Portugal"]
for i in lst:
    print(i)
```

```python
# 12. Iterate with index
#
# Using enumerate(), print:
countries = ["Netherlands" ,"Brazil" , "France" , "Argentina" , "Portugal"]

for index,countries in enumerate(countries):
    print(index,countries)
```

```python
# 13. Generate squares using list comprehension
#
# Create:
#
# [1, 4, 9, 16, ..., 100]

[i**2 for i in range(1,11)]
```

```python
# 14. Filter even numbers using list comprehension
#
# From 0â€“30, generate only even numbers.

[i for i in range(0,31) if i%2==0]
```

```python
# 16. Convert strings to uppercase
#
# Given:
#
# names = ["shravan", "ayush", "arjun"]
names = ["shravan", "ayush", "arjun"]
[ names.upper() for names in names  ]
```

```python
# 17. Generate pairs
#
# Given:
#
# a = [1,2,3]
# b = ["x", "y", "z"]

a = [1,2,3]
b = ["x", "y", "z"]
[(i,j) for i in a for j in b]
```

```python
# 18. Multiplication Table (nested list comprehension)
#
# Create a 5Ã—5 multiplication table like:
#
# [
#  [1,2,3,4,5],
#  [2,4,6,8,10],
#  ...
# ]


tables = [[i*j for i in range(1,6)] for j in range(1,6)]
print(tables)
```

```python
# 19. Flatten a 2D List (Nested List Comprehension)
#
# Given:
#
# matrix = [[1,2,3], [4,5,6], [7,8,9]]

matrix = [[1,2,3], [4,5,6], [7,8,9]]
flatten_matrix = [j for i in matrix for j in i]
print(flatten_matrix)
```
