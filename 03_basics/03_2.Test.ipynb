```python
# Part 5 — Nested Tuples
#
# Given the nested tuple:
#
# nt = ((10,20,30), ("a","b","c"), (True, False))
#
#
# Print the value "b"
#
# Print 30
#
# Print True
#
# Iterate over the nested tuple and print all items in a single line.

nt = ((10,20,30), ("a","b","c"), (True, False))
print(nt[1][1])
print(nt[0][-1])
print(nt[-1][0])
print("------------------------------------------")
for sub_tuple in nt:
    for index in sub_tuple:
        print(index , end=" ")
```

```python
# Find the largest and smallest element
#
# values = (15, 88, 23, 41, 7, 90)

## we will convert the tuple to list, and then we will sort the list and then we will convert the list into
## tuple and will print the first and the last index
val =  (15, 88, 23, 41, 7, 90)
lst = list(val)
lst.sort()
print(f"Smallest : {lst[0]} , Largest :  {lst[-1]}")

```

```python
# Remove duplicates from a tuple
#
# tup = (1,2,3,1,2,4,5,1)
#
#
# Output should be:
#
# (1,2,3,4,5)

tup = (1,2,3,1,2,4,5,1)
lst = list(tup)
for item in tup:
    if lst.count(item)>1:
        lst.remove(item)
# new_tup = tuple(sorted(lst))
lst.sort()
new_tup = tuple(lst)
print(new_tup)

# Logic: we are taking a tuple tup, and we are converting it into a list as shown , then we are traversing
#        over the tuple tup , and we kept a condition that if the lst which has the tup elements if the
#        item count is greater than 1 or repeated , then we have to remove the first occurrence of the
#        element for which we are using the .remove() method and we are removing the repeated occurrence of
#        item and after doing that the list will be unsorted , then inorder to sort it and convert it into a
#        tuple we are using tuple(sorted(lst))
#
#




```

```python
# Count total vowels in this tuple of strings:
#
# words = ("apple", "banana", "grapes", "kiwi")

words = ("apple", "banana", "grapes", "kiwi")
vowels = "aeiouAEIOU"
count = 0

# we require three variables one is words which is actual tuple, second is vowels and third is the count
# which increments if the vowel is found

for i in words:
    for j in i:
        if j in vowels:
            count = count+1
print("No of vowels=",count)

```
