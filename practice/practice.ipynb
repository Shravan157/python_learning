```python

# ============================================================================
# LEVEL 2: MODIFICATION & METHODS
# ============================================================================

print("\n" + "="*60)
print("LEVEL 2: MODIFICATION & METHODS")
print("="*60)

# Task 2.1: Given this dictionary, update the price to 899 and add a new key 'rating' with value 4.5
product = {"name": "Laptop", "brand": "Dell", "price": 799}
# Write your code here:
product['price'] = 899
product['rating'] = 4.5

print("Task 2.1 - Updated Product:")
print(product)
print()

# Task 2.2: Use get() method to safely access 'warranty' key. If not present, return "No warranty info"
# Write your code here:
product.get('warranty' , "No warranty info")

# Task 2.3: Remove the 'rating' key from product dictionary (if exists)
# Write your code here:
del product['rating']
print(product)

# Task 2.4: Create a copy of the product dictionary and modify the copy without affecting original
# Write your code here:
product_detail_copy = product.copy()
product_detail_copy['name'] = 'Desktop'
print(product_detail_copy)
print(product)
```

```python
# LEVEL 3: ITERATION
# ============================================================================

print("\n" + "="*60)
print("LEVEL 3: ITERATION")
print("="*60)

scores = {"Math": 85, "Science": 92, "English": 78, "History": 88, "Art": 95}

# Task 3.1: Print all subject names (keys)
print("Task 3.1 - All Subjects:")
# Write your code here:
for key , val in scores.items():
    print(f"{key}")


# Task 3.2: Print all scores (values)
print("\nTask 3.2 - All Scores:")
# Write your code here:
for key,val in scores.items():
    print(f"{val}")

# Task 3.3: Print each subject and its score in format "Subject: Score"
print("\nTask 3.3 - Subject-Score Pairs:")
# Write your code here:
for key,val in scores.items():
    print(f"{key}:{val}")

# Task 3.4: Calculate and print the average score
print("\nTask 3.4 - Average Score:")
# Write your code here:
avg_score = sum(scores.values()) / len(scores.values())
print(avg_score)
```
